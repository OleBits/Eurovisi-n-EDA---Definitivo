### Análisis de eurovisión

### TEMA

* Definir las tendencias en las propuestas que se mandan a eurovisión en los últimos años, cuáles son los perfiles que más puntos han obtenido, diferenciando tanto los puntos del jurado como del televoto.
* Definir las tendencias de votos a lo largo  de los años y ver si estas cambian con el cambio de sistema de votos.

Consideraciones específicas del tema:

* Es un concurso organizado por la Unión Europea de la Radiodifusión (UER), donde ser europeo no es un requisito, por lo que hay paises no europeos que participan, al igual que hay paises europeos que no participan siempre.
* Se realizan dos semifinales eliminatorias y una final. Los 5 paises que más aportan económicamente a la UER son conocidos como "Big five" y tienen clasificación directa a la final, estos son: Alemania, España, Francia, Italia y Reino Unido. Italia no participó entre los años 2000 y 2010.
* El número de participantes no es fijo cada año, el número de finalistas tampoco, por lo que el valor de puntos entre una edición a otra es relativo.
* No se celebró en 2020 a causa de la pandemia
* El sistema de votación es el siguiente:
    * Cada país otorga de 1 a 8, 10 y 12 puntos a sus 10 países favoritos.
    * Hasta 1975 el formato no es uniforme.
    * Desde 1975 hay estos "12 points", voto de jurado.
    * Desde 1997 solo hay televoto.
    * El televoto es de pago, no se puede votar al mismo país en el que se está, siempre dentro de la UER.
    * A partir de 2009 el voto es mixto jurado-televoto, donde tienen un peso del 50%, ponderando el televoto y sumándolo.
     * A partir de 2016 el voto del jurado y del televoto está separado, por lo que se repartirán el doble de puntos que con el método anterior.

### HIPÓTESIS

* Creo que habrá ciertas tendencias que serán predominantes (como el estilo musical) y que serán las tendencias que más se repiten entre las propuestas mejor  puntuadas de cada año.
* Habrá paises que siempre reciben puntos de otros de su misma región (voto vecinal) y que el uso del jurado reducirá esta tendencia.
